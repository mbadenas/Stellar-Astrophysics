###  Adiabatic Exponents

$$
\Gamma_{1}\equiv-\left(\frac{\partial\ln P}{\partial\ln V}\right)_{s}$$

$$
 \frac{\Gamma_{2}}{\Gamma_{2}-1}\equiv\left(\frac{\partial\ln P}{\partial\ln T}\right)_{s}$$
 
 
$$1-\Gamma_{3}=\left(\frac{\partial\ln T}{\partial\ln V}\right)_{s}$$
 

```python
beta = np.arange(0.00001,1,0.01)

num1 = 32-24*beta-3*(beta**2)
den1 = 24-21*beta
gamma1 = num1/den1

num2 = 32-24*beta-3*(beta**2)
den2 = 24-18*beta-3*(beta**2)
gamma2 = num2/den2

num3 = 32-27*beta
den3 = 24-21*beta
gamma3 = num3/den3
```

```python
fig = plt.figure()
ax = fig.add_subplot(111)

plt.axhline(y=(5.0/3.0), color='k', linestyle='--', linewidth=0.3)
plt.axhline(y=(4.0/3.0), color='k', linestyle='--', linewidth=0.3)
plt.text(max(beta),(5.0/3.0)+0.01,'$\Gamma=5/3$')        
plt.text(max(beta),(4.0/3.0)+0.01,'$\Gamma=4/3$')  

plt.text(0.9,1.6,'Gas')
plt.text(0.2,1.3,'Radiation')

ax.plot(beta,gamma1, label = '$\Gamma_{1}$')
ax.plot(beta,gamma2, label = '$\Gamma_{2}$')
ax.plot(beta,gamma3, label = '$\Gamma_{3}$')

ax.set_xlim([1, 0])
ax.set_ylim([1.2, 1.8])

ax.set_xlabel(r'$\beta$')
ax.set_ylabel(r'$\Gamma$')
ax.set_title('Adiabatic Exponents for a Mixture of Gas and Radiation')
ax.legend()
```

For a gas, $\beta = 1$, and  $\Gamma_{1} = 5/3 = \gamma$ (as we had seen for an ideal gas). For radiation, $\beta=0$ and $\Gamma_{1} =4/3$.
